```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd

df = pd.read_csv('C:\\Users\\Haine\\OneDrive\\Escritorio\\Proyecto_2\\Data\\bank_cleaned.csv')  # Reemplaza con el nombre real del archivo
df

print(df.head())          # Ver las primeras filas
print(df.columns.tolist())  # Ver los nombres de las columnas


# Supongamos que df ya está cargado

# Variables predictoras (X) y variable objetivo (y)
X = df.drop('contact', axis=1)
y = df['default']


# Definimos las columnas numéricas y categóricas
numeric_features = ['age', 'balance']
categorical_features = ['job', 'campaign']

# Asegúrate de que estas columnas existen en df:
print(X.columns)

# Preprocesador
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ]
)

# Pipeline con preprocesamiento + modelo
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

# División de datos
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Entrenar el pipeline
pipeline.fit(X_train, y_train)

# Acceso a preprocesador ya entrenado
preprocessor_fitted = pipeline.named_steps['preprocessor']
print(preprocessor_fitted.transformers_)


```
